```python
data.head(1)
```

```python
data.info()
```

```python
data = data[["content_duration","num_subscribers"]]
```

```python
plt.scatter(data["content_duration"] , data["num_subscribers"] , color="green")
plt.title("Course content duration and number of subscribers ")
plt.xlabel("Content Duration")
plt.ylabel("No. of Subscribers")
plt.show()
```

```python
data.describe()
```

```python
from sklearn import linear_model
```

```python
train = data[:(int((len(data)*0.8)))]
test = data[(int((len(data)*0.8))):]
```

```python
regr = linear_model.LinearRegression()
train_x = np.array(train[["content_duration"]])
train_y = np.array(train[["num_subscribers"]])
regr.fit(train_x,train_y)
```

```python
print ("coefficients : ",regr.coef_) #Slope
print ("Intercept : ",regr.intercept_) #Intercept
```

```python
##Linear Regression representation using scatter plot
plt.scatter(train["content_duration"], train["num_subscribers"], color='blue')
plt.plot(train_x, regr.coef_*train_x + regr.intercept_, color='red')
plt.title(" Linear Regression-Course content duration and number of subscribers ")
plt.xlabel("Content Duration")
plt.ylabel("No.of Subscribres")
```

```python
def get_regression_predictions(input_features,intercept,slope):
    predicted_values = input_features*slope + intercept
    return predicted_values
```

```python
no_subs = 2
nosubs = get_regression_predictions(no_subs,regr.intercept_[0],regr.coef_[0][0])
"num_subscribers:",nosubs
```

```python
from sklearn.metrics import r2_score 
test_x = np.array(test[['content_duration']]) 
test_y = np.array(test[['num_subscribers']]) 
test_y_ = regr.predict(test_x)
print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Mean sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y)** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```
